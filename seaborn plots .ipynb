```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
np.random.seed(42) #to get the same random dataset each time.

df = pd.DataFrame({
    'Sales': np.random.randint(200, 500, 50), #random number between 200 to 500 for 50 times
    'Profit': np.random.randint(50, 150, 50),    #random number between 50 to 150 for 50 times
    'Expenses': np.random.randint(100, 300, 50),     #random number between 100 to 300 for 50 times
    'Category': np.random.choice(['A', 'B', 'C'], 50),     #randomly choose A or B or C for 50 times
    'Region': np.random.choice(['North', 'South'], 50)     #randomly choose North or South for 50 times
})

df
```

```python
#  Distribution plots → histplot, jointplot, pairplot, rugplot, kdeplot
# Categorical plots → catplot (factorplot), boxplot, violinplot, stripplot, swarmplot, barplot, countplot
# Matrix plot → heatmap
```

```python
#Distribution Plots ------------ show how data points are spread
#   1. displot() — figure-level function (deprecated → histplot / displot)

sns.histplot(df['Sales'], kde=True, color='red')
plt.title('Distribution of Sales')
plt.show()
```

```python
#   2. jointplot()

sns.jointplot(x='Sales', y='Profit', data=df, kind='scatter')
sns.jointplot(x='Sales', y='Profit', data=df, kind='kde', fill=True)
```

```python
#pairplot
sns.pairplot(df, hue='Category')
```

```python
#   3. kdeplot() — kernel density estimation curve
sns.kdeplot(data=df, x='Sales', fill=True, color='purple')
plt.show()
```

```python
#   4. rugplot() — marks actual data points along an axis
sns.rugplot(data=df, x='Sales', height=0.1, color='red')
plt.show()
```

```python
#Categorical Data Plots---used when at least one of your variables is categorical(like "A", "B", "North", "South")
#   1. factorplot (deprecated → catplot)

sns.catplot( data=df, x='Category', y='Sales', kind='bar')
```

```python
#    2. boxplot 
sns.boxplot(data=df, x='Category', y='Sales', hue='Category', palette='Set2', legend=True)
```

```python
#violinplot
sns.violinplot(x='Category', y='Sales', hue='Category', data=df, palette='Set3', legend=True)
```

```python
#stripplot
sns.stripplot(x='Category', y='Sales', data=df, jitter=True, color='purple')
```

```python
#swarmplot
sns.swarmplot(x='Category', y='Sales',hue='Category', data=df, palette='husl')
```

```python
#barplot
sns.barplot(x='Category', y='Sales',hue= 'Category', data=df, ci='sd', palette='muted')
```

```python
#countplot
sns.countplot(x='Category',hue='Category', data=df, palette='pastel')
```

```python
#Matrix Plots
# Heatmap
corr = df[['Sales', 'Profit', 'Expenses']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
