```python
#Linear Regression Multiple variables
#equation---> y=m1*X1 + m2*X2 + m3*X3 +C
#X1,X2,X3 are multiple independent variables
#m1,m2,m3 are slope/gradient/coefficient  and c is intercept
```

```python
import pandas as pd
from sklearn import linear_model
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\k4kas\OneDrive\Documents\Downloads\Insurance.csv')
```

```python
df
```

```python
#firstly analyze data (like continuity and missing values)
#missing values replace with mean
 
#to find mean of particular colummn
mean=df.Height.mean()
```

```python
#to replace missing values with mean
df.Height.fillna(mean)
```

```python
#but still mean is not in our data set
df
```

```python
#to replace missing values in dataset
df.Height =df.Height.fillna(mean)

#display dataset
df
```

```python
#reg is object of linear regression model
reg=linear_model.LinearRegression()
```

```python
#to train regression model
reg.fit(df[['Age','Height','Weight']],df[['Premium']])
```

```python
#now predict
reg.predict([[27,165.56,60]])
```

```python
#to find slope
reg.coef_
```

```python
#to find intercept
reg.intercept_
```
