```python
# Discretization & Binning in pandas
#Used to convert continuous data into discrete bins
import pandas as pd

ages = [15, 22, 45, 60, 75]
bins = [0, 18, 35, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Senior']

ad =pd.cut(ages, bins, labels=labels)
print(ad)

#to find max and min
ad.min(), ad.max()
```

```python
#ploting

import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Score': [85, 90, 88, 92]
})
print(df)

# 1.Univariate (one variable)-----Focuses on the distribution of a single variable.

#histogram
df['Age'].plot.hist()

#Box Plot (Whisker Plot)
df['Age'].plot.box()

#KDE Plot (smooth curve of distribution)
df['Age'].plot.kde()
```

```python
# 2.Bivariate (two variables) ---------- Shows the relationship between two variables

#scatter plot
df.plot.scatter(x='Age', y='Score')

#Correlation Coefficient
df.corr(numeric_only=True)
```

```python
#Multivariate (three or more variables)

import pandas as pd
import seaborn as sns

data = {
    'Age': [23, 30, 27, 25, 35],
    'Height': [165, 175, 160, 170, 180],
    'Weight': [55, 75, 50, 65, 85],
    'Score': [80, 70, 85, 78, 90]
}
df = pd.DataFrame(data)

# Pairplot - visualizes relationships between all variables
sns.pairplot(df)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Year': [2018, 2019, 2020, 2021, 2022],
    'Sales': [25000, 27000, 30000, 28000, 32000],
    'Salary': [40000, 42000, 45000, 47000, 50000],
    'Age': [25, 28, 35, 40, 45],
    'Profit': [5000, 7000, 8000, 7500, 9000],
    'Customers': [200, 230, 250, 240, 260]
})

print(df)
```

```python
#Pandas Plotting Functions

#line plot-----default one
df.plot.line(x='Year', y='Sales')
```

```python
#Area Plot
df.plot.area()
```

```python
#Bar Plot (Vertical)
df.plot.bar()
```

```python
#Bar Plot (Horizontal)
df.plot.barh()
```

```python
#Density Plot (KDE)
df.plot.kde()
```

```python
#Histogram
df['Age'].plot.hist()
```

```python
#Scatter Plot
df.plot.scatter(x='Age', y='Salary')
```

```python
#Box Plot (Whisker Plot)
df.plot.box()
```

```python
# Hexbin Plot
df.plot.hexbin(x='Age', y='Salary')
```

```python
# Pie Chart
import matplotlib.pyplot as plt

#plot pie chart for Sales
df['Sales'].plot.pie(
    labels=df['Year'],     #labels
    autopct='%1.1f%%',     # Show percentage 
)
plt.title("Sales Distribution by Year")
plt.show()
```
