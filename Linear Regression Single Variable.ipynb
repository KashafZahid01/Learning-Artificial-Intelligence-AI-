```python
#Linear Regression-----based on supervised Learning used to predict continuous/real or numerical values
#Linear Equation--------> y=mx+c
#Linear Regression single variable
```

```python
import pandas as pd
import seaborn as sns
from sklearn import linear_model
```

```python
df=pd.read_csv(r'C:\Users\k4kas\OneDrive\Documents\Downloads\Insurance.csv')
```

```python
df
```

```python
#to draw best fit line
sns.lmplot(x='Age',y='Premium',data=df)
```

```python
#linear regression method
reg=linear_model.LinearRegression()
```

```python
#to train regression model
reg.fit(df[['Age']],df[['Premium']])
```

```python
#now model is ready to predict
#predict premium for 21 age
reg.predict([[21]])
```

```python
#predict premium for 50 age
reg.predict([[50]])
```

```python
#to get slope of equation (y=mx+c) which is "m".
reg.coef_
```

```python
#to get intercept of equation (y=mx+c) which is "c".
reg.intercept_
```
