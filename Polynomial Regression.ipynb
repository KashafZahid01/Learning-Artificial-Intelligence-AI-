```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r'C:\Users\k4kas\OneDrive\Documents\Downloads\office_sheet.csv')
```

```python
dataset
```

```python
#to store independent variable( Level) in x
x= dataset.iloc[:,1:2]
```

```python
#its in series
x
```

```python
#to get x in array form
x= dataset.iloc[:,1:2].values

x
```

```python
#similarly for dependent variable(Salary)
y= dataset.iloc[:,2:3].values
y
```

```python
#to visualize  dataset
plt.scatter(x,y)
```

```python
#for best fit line
#as best line line passes thorugh only one point means error rate is higher
sns.lmplot(x='Level',y='Salary',data=dataset)
```

```python
from sklearn import linear_model
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
#its not better result as error rate is higher
reg.predict([[6.5]])
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2)
```

```python
#fit independent variable
x_poly=poly.fit_transform(x)
```

```python
reg2= linear_model.LinearRegression()
```

```python
reg2.fit(x_poly,y)
```

```python
#result after using PolynomialFeatures
reg2.predict(poly.fit_transform([[6.5]]))
```
